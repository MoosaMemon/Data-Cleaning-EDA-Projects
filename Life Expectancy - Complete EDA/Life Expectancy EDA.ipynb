## **Importing libraries and reading data:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("Life Expectancy Data.csv")
```

## **Sanity check and data cleaning:**

```python
df.shape
```

```python
df.head()
```

```python
# No duplicate values found in the dataset
df.duplicated().sum()
```

```python
for i in df.select_dtypes(include="object").columns:
    print(df[i].value_counts())
    print("---------------------------------")
```

```python
df["Status"] = df["Status"].replace({"Developing": 0, "Developed": 1}, regex=True)
```

```python
df.rename(columns={"Status": "isDeveloped"}, inplace=True)
```

```python
# missing values present in the dataset
df.isnull().sum()/df.shape[0]*100
```

```python
# i will treat the missing values using knn imputer, one can also fill the missing values with either mean, median, or mode
# knn imputer takes average of the nearest value and fill the missing value with the average of the missing value:

from sklearn.impute import KNNImputer
impute = KNNImputer()

for i in df.select_dtypes(include="number").columns:
    df[i] = impute.fit_transform(df[ [i] ])
```

```python
df.tail()
```

```python
df.info()
```

```python
df.columns
```

```python
# converting columns to their appropriate datatypes
cols_to_conv = ["Year", "isDeveloped", "infant deaths", "Population", "under-five deaths ", "Adult Mortality", "Diphtheria ", "Hepatitis B", "Measles ", "Polio"]
for i in cols_to_conv:
    df[i] = df[i].astype(int)
```
